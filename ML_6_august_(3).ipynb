# K mean

```python
df=pd.read_csv(r"C:\Users\lab1\Downloads\income.csv")
df
```

```python
plt.scatter(df.Age,df["Income($)"])
plt.xlabel("Age")
plt.ylabel("Income")
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 3)
```

```python
y = km.fit_predict(df[["Age","Income($)"]])
y
```

```python
df["Clusters"]=y
df
```

```python
km.cluster_centers_
```

```python
df1=df[df.Clusters==0]
df2=df[df.Clusters==1]
df3=df[df.Clusters==2]
```

```python
df1
```

```python
plt.scatter(df1.Age,df1["Income($)"],color="green")
plt.scatter(df2.Age,df2["Income($)"],color="blue")
plt.scatter(df3.Age,df3["Income($)"],color="black")
plt.legend()
plt.xlabel("Age")
plt.ylabel("Income")
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
s = MinMaxScaler()
```

```python
s.fit(df[["Income($)"]])
```

```python
df["Income($)"]=s.transform(df[["Income($)"]])
```

```python
df.head()
```

```python
s.fit(df[["Age"]])
```

```python
df["Age"]=s.transform(df[["Age"]])
```

```python
df.head()
```

```python
km=KMeans(n_clusters=3)
ys=km.fit_predict(df[["Age","Income($)"]])
ys
```

```python
df["Clusters"]=ys
df
```

```python
df1=df[df.Clusters==0]
df2=df[df.Clusters==1]
df3=df[df.Clusters==2]
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
plt.scatter(df1.Age,df1["Income($)"],color="green")
plt.scatter(df2.Age,df2["Income($)"],color="blue")
plt.scatter(df3.Age,df3["Income($)"],color="black")
plt.legend()
plt.xlabel("Age")
plt.ylabel("Income")

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="purple",marker="*")
```

```python
se=[]
k_r=range(1,10)
for k in k_r:
    km=KMeans(n_clusters=k)
    km.fit(df[["Age","Income($)"]])
    se.append(km.inertia_)
```

```python
se
```

```python
plt.plot(k_r,se)
plt.xlabel("K")
plt.ylabel("sum of squared error")
```
